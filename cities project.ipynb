# Merge Data sets

```python
cities.head(5)
```

```python
world_cities.head(5)
```

```python
#select some features on cities data set and drop the unneeded column
cities1 = cities.drop(columns=['Id','SmartCity_Index_relative_Edmonton'])
cities1
```

```python
#to make merging should we have a same coulmn 
continent2=clean_continent.rename(columns={'Country_Name': 'Country'})
```

```python
continent2
```

```python
#merge cities and continent on the column country 
firstMerge=cities1.merge(continent2, how='inner', on=['Country'])
firstMerge
```

```python
clean_data=firstMerge.drop(columns=['Continent_Code','Continent_Code','Two_Letter_Country_Code','Three_Letter_Country_Code','Country_Number'])
clean_data
```

```python
world_cities2=world_cities.rename(columns={'city':'City'})
world_cities2                                  
                                           
```

```python
#drop unneeded column befor final merging
world_cities3=world_cities2.drop(columns=['city_ascii','id','country'])
world_cities3
```

```python
final_merge=clean_data.merge(world_cities3, how='inner', on=['City'])
final_merge
```

```python
#as shown there is some repeated data and wrong so drop it
final_merge1= pd.DataFrame(final_merge)

rows_to_drop  = [1,2,3,5,9,14,15,16]
final_merge1.drop(rows_to_drop, inplace=True)
final_merge1
```

```python
#as it clear our final dataset is clean befor visualization

final_merge1.isnull().sum()
```

```python
heatmap_data=final_merge.drop(columns=['Country','Continent_Name','lat','lng','iso2','iso3'])
heatmap_data
```

```python
#as shown there is some repeated data and wrong so drop it
heatmap_data1= pd.DataFrame(heatmap_data)
rows_to_drop = [1, 2,3,5,9,14,15,16]
heatmap_data1.drop(rows_to_drop, inplace=True)
heatmap_data1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

heatmap_datasorted = heatmap_data1.sort_values(by='SmartCity_Index', ascending=False)

heatmap_datasorted.set_index('City', inplace=True)
top_10_cities = heatmap_datasorted.head(10).drop(columns=['SmartCity_Index'])
plt.figure(figsize=(10, 6))  
sns.heatmap(top_10_cities, cmap='Blues', annot=True, fmt=".1f") 
plt.title('Heatmap of Cities')
plt.show()
```

```python
cities_with_location=top_10_cities.merge(world_cities3, how='inner', on=['City'])
cities_with_location
```

```python
#delet wrong data after merg
deleted=[1,4,5,6,9,14,15,16]
cities_with_location.drop(deleted, inplace=True)
cities_with_location
```

```python
import folium
```

```python
mymap = folium.Map(location=[0, 0], zoom_start=2)

# Add markers for each city
for index, row in cities_with_location.iterrows():
    folium.Marker(location=[row['lat'], row['lng']], popup=row['City']).add_to(mymap)

# Save the map to an HTML file
mymap.save("cities_map.html")
```
